```python
id_vars = list(modified_data.columns)[0:8]
value_vars = list(modified_data.columns)[8:]
```

```python
dataset_melted = modified_data.melt(id_vars=id_vars,value_vars=value_vars,var_name='Question + Subquestion', value_name='Answer')
```

```python
dataset_melted
```

```python
question_imported = pd.read_excel(r"C:\Users\Akash kumar Ram\OneDrive\Documents\Data_Analysis_Project\Survery_monkey_data\Data_ Survey Monkey Output.xlsx",sheet_name="Question")
```

```python
question_imported.drop(["Raw Question","Raw Subquestion"],axis=1,inplace=True)
```

```python
question_imported.dropna(inplace=True)
```

```python
Merged_data = pd.merge(left = dataset_melted,right=question_imported, how="left" ,left_on="Question + Subquestion",right_on="Question + Subquestion" )
```

```python
Merged_data
print("Original_data",len(dataset_melted))
print("Merged_data",len(Merged_data))
Merged_data
```

```python
respondents = Merged_data[Merged_data['Answer'].notna()]
respondents = respondents.groupby('Question')['Respondent ID'].nunique().reset_index()
respondents.rename(columns={'Respondent ID':"Respondents"},inplace=True)
```

```python
respondents
```

```python
merged2 = pd.merge(left=Merged_data,right=respondents,how='left',left_on='Question',right_on ='Question')
                   
                  
```
